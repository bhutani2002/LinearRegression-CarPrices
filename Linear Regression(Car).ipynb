Model Building

```python
#train_test_split is a function in Sklearn model selection for splitting data arrays into two subsets: for training data and for testing data
X_Train,X_Test,Y_Train,Y_Test=train_test_split(df,Y,train_size=0.75,random_state=42)
```

```python
#sklearn.preprocessing is a package
"""StandardScaler--""" """standardizes a feature by subtracting the mean and then scaling to unit variance. 
   Unit variance means dividing all the values by the standard deviation.
   StandardScaler makes the mean of the distribution 0"""

#fit_transform : used on the training data so that we can scale the training data and also learn the scaling parameters of that data
```

```python
from sklearn.preprocessing import StandardScaler
standard=StandardScaler()
X_Train=standard.fit_transform(X_Train)
X_Test=standard.fit_transform(X_Test)

print(X_Train)
print("----"*50)
print(X_Test)
```

```python
#Definition: Linear Regression is a machine learning algorithm based on supervised learning. It performs a regression task. Regression models are target prediction value based on independent variables. 
#Advantages
#   1.Linear Regression is simple to implement and easier to interpret the output coefficients.
#   2.When you know the relationship between the independent and dependent variable have a linear relationship, this algorithm is the best to use because of itâ€™s less complexity to compared to other algorithms.
#   3.Linear Regression is susceptible to over-fitting but it can be avoided using some dimensionality reduction techniques, regularization (L1 and L2) techniques and cross-validation

#Disadvantages
#   1.On the other hand in linear regression technique outliers can have huge effects on the regression and boundaries are linear in this technique.
#   2.Diversely, linear regression assumes a liession also looks at a relationship between the mean of the dependent variables and the independennear relationship between dependent and independent variables. That means it assumes that there is a straight-line relationship between them. It assumes independence between attributes.
#   3.But then linear regrt variables. Just as the mean is not a complete description of a single variable, linear regression is not a complete description of relationships among variables.
```

```python
from sklearn.linear_model import LinearRegression
Regression_lr=LinearRegression()
Regression_lr.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#Definition: LightGBM is a gradient boosting framework that uses tree based learning algorithms. It is designed to be distributed and efficient.
#Advantages: Faster training speed and higher efficiency. Lower memory usage.
#Disadvantages:using this LightGBM is its narrow user base   
```

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
Regression_lgbm=LGBMRegressor()
Regression_lgbm.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#Definition: Stochastic Gradient Descent. Stochastic Gradient Descent (SGD) is a simple yet very efficient approach to fitting linear classifiers .
#Advantages: It is easier to fit in the memory due to a single training example being processed by the network.It is computationally fast as only one sample is processed at a time.
#Disadvantages: Due to frequent updates, the steps taken towards the minima are very noisy. This can often lean the gradient descent into other directions.
```

```python
from sklearn.linear_model import SGDRegressor
Regression_sgdr=SGDRegressor()
Regression_sgdr.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#Definition: learns a linear function in the space induced by the respective kernel and the data
#Advantages: Computational simplicity
#Disadvantages: no concept of â€œsupport vectors.â€This makes a difference at prediction time
```

```python
from sklearn.kernel_ridge import KernelRidge
Regression_kr=KernelRidge()
Regression_kr.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#Definition: A statistical model where you use probability to represent all uncertainty within the model
#Advantages: allows a natural mechanism to survive insufficient data or poorly distributed data by formulating linear regression using probability distributors rather than point estimates
#Disadvantages: It's not great when you don't have a regression problem, or if a linear model does not work well
```

```python
from sklearn.linear_model import BayesianRidge
Regression_br=BayesianRidge()
Regression_br.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#Definition: Support Vector Regression is a supervised learning algorithm that is used to predict discrete values. Support Vector Regression uses the same principle as the SVMs. The basic idea behind SVR is to find the best fit line. 
#Advantages: SVR is robust to the outliers.Decision model can be easily updated.It can improve the prediction accuracy by measuring the confidence in classification.
#Disadvantages: there's nothing really to fall back on in order to estimate the uncertainty contained in the results SVR provides.
```

```python
from sklearn.svm import SVR
Regression_svr=SVR()
Regression_svr.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#Definition: A random forest is a machine learning technique that's used to solve regression and classification problems. It utilizes ensemble learning, which is a technique that combines many classifiers to provide solutions to complex problems. A random forest algorithm consists of many decision trees
#Advantages: it reduces overfitting problem in decision trees and also reduces the variance and therefore improves the accuracy
#Disadvantages: Complexity, Longer Training Period
```

```python
from sklearn.ensemble import RandomForestRegressor
Regression_rfr=RandomForestRegressor()
Regression_rfr.fit(X_Train,Y_Train["Selling_Price"])
```

```python
#List of the models
regression_list=[Regression_lr,Regression_br,Regression_kr,Regression_lgbm,Regression_rfr,Regression_sgdr,Regression_svr]
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

CALCULATE ACCURACY, MEAN SQUARED ERROR AND PLOT DISPLOT GRAPH

```python
#displot: This function provides access to several approaches for visualizing the univariate or bivariate distribution of data.
for regression in regression_list:
    Y_pred=regression.predict(X_Test)
    Y_pred=Y_pred.reshape(76,1)
    rmse_value=mean_squared_error(Y_Test,Y_pred,squared=False)
    sns.displot(Y_Test-Y_pred)
    # plt.plot(X_Test[:,0],Y_pred)
    print("Regression Model:\n {}".format(regression))
    print("     Accuracy:",r2_score(Y_pred,Y_Test)*100)
    print("     Mean squared error: ",rmse_value)
    print("-"*50)
    # print(regression.summary())
```
